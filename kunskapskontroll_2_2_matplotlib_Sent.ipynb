### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# In Matplotlib, there are two main ways to make plots:
#1. Using Pyplot where you use functions like plt.plot() and plt.title() directly. It can be used to generate simple plots in quick and easy way.
#2. Object-Oriented (OO) way: You first create objects (Figure and Axes) and then call methods on them, like ax.plot() and ax.set_title(). It is used for more complex plots and gives you more control.

# The Object-Oriented style is recommended by the official Matplotlib guide because it is cleaner, more flexible, and better for larger projects.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure = the page or canvas where everthing is drawn. It can contain one or more plots.
# Axes = one single plot in the figure (canvas). You can have multiple Axes in one Figure.
# Axis = The x-axis or y-axis inside an Axes.
# Artist = everything you see: lines, text, labels, points, etc.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# Lists or NumPy arrays used to represent the x- and y-values
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Create a figure containing 1 row, 2 columns and two axes
# Constrained adjusts the elements of the plots to not overlap (instead of using plt.tight_layout)
#fig, ax = plt.subplots(1, 2, figsize=(10, 4), layout='constrained') 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), layout='constrained')

# Create plot 1 using pyplot approach
plt.sca(ax1) # select the first subplot/axes
plt.plot(x, y)
plt.title("Plot of y = x² (Using Pyplot approach)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Create plot 2 with OO approach
ax2.plot(x, y)
ax2.set_title("Plot of y = x² (Using OO approach)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.grid(True)

# Show plots in figure
plt.show()

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Colors for each bar
colors = ['blue', 'green', 'yellow', 'orange', 'red']
```

```python
# Create figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), layout='constrained')

# Plot 1: Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color='blue')
ax1.set_title("My scatter plot")
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax1.grid(True)

# Plot 2: Bar plot
ax2.bar(names, values, color=colors)
ax2.set_title("Number of fruits")
ax2.set_ylabel("Quantity")

# Show the plots in figure
plt.show()
```

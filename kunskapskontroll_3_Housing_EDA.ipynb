## Prerequisites

```python
# Import NumPy, Pyplot och Pandas. 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Import dataset (LOAD)
housing_original = pd.read_csv("housing.csv")
```

## Clean-up of the data (TRANSFORM)

```python
housing_transformed = housing_original.dropna() # Drop rows with missing values (save the original data set before dropping rows)
housing_transformed.info() # show info about data set after dropping rows
```

## Visualisations

```python
housing_transformed.describe() # common statistics of the data set
```

```python
print("Medelvärde för median_house_value:", round(housing_transformed["median_house_value"].mean(), 0))
print("Minsta värde:", housing_transformed["median_house_value"].min())
print("Högsta värde:", housing_transformed["median_house_value"].max())
```

### Samband mellan inkomst och huspris
##### Varje punkt i diagrammet visar ett geografiskt område. X-axeln visar hushållets inkomst och y-axeln visar huspris. 
##### Diagrammet visar ett tydligt samband mellan inkomst och bostadspriser: ju högre medianinkomst desto högre bostadspris. Det finns ett särskilt kluster av datapunkter i intervallet mellan 2-6 i inkomst och upp till 300 000 i bostadspris. Vid inkomst över 6 börjar bostadspriserna öka. Att många punkter ligger i toppen på 500,000 i inkomst tyder på att det finns ett tak för huspriser i datasetet. Detta diagram ger en bra visuell grund för att bevisa att det finns relation mellan bostadspriser och inkomst vilket är en viktig insikt för att förutspå framtida köp av hus i området.
#####

```python
plt.figure(figsize=(7, 5)) # Create a new figure
plt.scatter(housing_transformed["median_income"], housing_transformed["median_house_value"], alpha=0.6,edgecolor='black') # Create a scatter plot, # edgecolor='black' adds a black outline to each point for better visibility
plt.title("Median house value and median income")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.grid(True)
plt.show() # Display the plot
```

### Antal sovrum i relation till inkomst eller huspris
##### Nedan diagram visar sambandet mellan antal sovrum i ett område och medianinkomst eller medianhuspriser. 
##### Antal sovrum verkar inte ha något starkt samband med varken inkomst eller huspriser. Det finns dock intressanta mönster som tyder på att det är andra faktorer än storlek (sovrum) som påverkar inkomst och pris mest. Antalet sovrum kan påverkas av andra faktorer än bara inkomst och huspris, t.ex. familjestorlek, bostadsutbud, eller hustyp. Vissa områden med många sovrum har ändå låg inkomst, vilket kan tyda på att det är stora men enkla/flerfamiljsboenden. När man bygger en prediktiv modell bör man därför kombinera antal sovrum med andra variabler som location, population, närhet till hav, etc
#####

```python
# Create a figure containing 1 row, 2 columns and two axes
# Constrained adjusts the elements of the plots to not overlap (instead of using plt.tight_layout)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), layout='constrained')

# Plot a scatter plot on the first axis (ax1)
# alpha=0.6 makes points semi-transparent; edgecolor='black' adds a black outlin
housing_transformed.plot(kind="scatter", x="total_bedrooms", y="median_income", alpha=0.6, edgecolor='black', ax=ax1)
ax1.set_title("Number of bedrooms vs Income")
ax1.set_xlabel("Number of bedrooms")
ax1.set_ylabel("Median income")
ax1.grid(True)

# Plot a second scatter plot on the second axis (ax2)
housing_transformed.plot(kind="scatter", x="total_bedrooms", y="median_house_value", alpha=0.6, edgecolor='black', ax=ax2)
ax2.set_title("Number of bedrooms vs median house value")
ax2.set_xlabel("Number of bedrooms")
ax2.set_ylabel("Median house value")
ax2.grid(True)

# Display the entire figure with both subplots
plt.show()
```

### Medelvärde för huspriser grupperat på närhet till havet
##### Diagrammet visar genomsnittliga bostadspriser beroende på hur nära havet bostadsområdena ligger.
##### Visualiseringen visar att hus nära havet generellt har ett högre genomsnittligt pris än hus längre in i landet

```python
# Group the data by 'ocean_proximity' and calculate the mean house value for each group
# Then sort the values in ascending order
mean_prices = housing_transformed.groupby("ocean_proximity")["median_house_value"].mean().sort_values()

# Plot the mean prices as a horizontal bar chart
mean_prices.plot(kind="barh", figsize=(8, 5), color="steelblue", edgecolor="black")
plt.title("Medelvärde för huspriser per område")
plt.xlabel("Medelvärde ($)")
plt.ylabel("Område")
plt.grid(axis="x")

# Add data labels (price values) next to each bar
for i, value in enumerate(mean_prices):
    plt.text(value + 5000, i, f"${value:,.0f}", va='center', fontsize=10)

plt.show() # Display the plot
```

### Sambandet mellan geografiskt läge, bostadspriser och befolkning i Kalifornien
##### Varje punkt representerar ett område. Färgen på punkten visar medianvärdet på bostäder i området (från blå = billig till röd = dyr). Storleken på punkten beror på befolkningen i det området.
##### Karta visar sambandet mellan geografiskt läge, bostadspriser och befolkning i Kalifornien. Områden längs kusten har högre bostadspriser (rödare färger). Dessa områden är även mer tätbefolkade, vilket syns på de större punkterna. Inlandet har generellt lägre bostadspriser och mindre befolkning. Det är tydligt att läget nära havet har både högre efterfrågan och pris på bostäder. Genom att kombinera karta, färg, storlek och data får man en mycket tydlig visuell översikt över bostadsmarknaden i området.

```python
import matplotlib.image as mpimg

# Load the image of California to be used as a map background
california_img=mpimg.imread('california.png')

# Create a scatter plot on a map: longitude vs latitude
# Point size represents population (larger = more people)
# Color represents median house value (color scale is 'jet')
# Alpha=0.4 makes points more visible
# No colorbar added yet (colorbar=False here)
ax = housing_transformed.plot(kind="scatter", x="longitude", y="latitude", figsize=(10,7), 
                  s=housing_transformed['population']/100, label="Population",
                  c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=False, alpha=0.4)

# Overlay the California map image onto the same plot
plt.imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5, cmap=plt.get_cmap("jet"))

# Set axis labels
plt.ylabel("Latitude")
plt.xlabel("Longitude")

# Create a custom colorbar using the same value range as the house prices
prices = housing_transformed["median_house_value"]
# Normalize the colorbar ticks and format them as $xxxk
tick_values = np.linspace(prices.min(), prices.max(), 11)
cbar = plt.colorbar(ticks=tick_values/prices.max())
cbar.ax.set_yticklabels(["$%dk"%(round(v/1000)) for v in tick_values], fontsize=14)
cbar.set_label('Median house value', fontsize=16)

# Add a legend for the population size and display the final map plot
plt.legend(fontsize=16)
plt.show()
```
